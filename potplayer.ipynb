```python
import pyautogui
import random
import time
from pynput import keyboard
import threading
```

```python
global keep
keep = True
def started():
#     點選播放
#     pyautogui.moveTo(36,1040)
#     pyautogui.click(button = 'left' )

    # 點選進度條 x軸範圍：0300 ~ 1750
    x = random.randint(6,10)
    avg_range = int(1460/x)
    global keep
    keep = True
    while keep:
        list=[]
        for i in range(x):
            num_1 = int(avg_range*(i+1))
            num_2 = int(avg_range*(i+2))
            num = random.randint(num_1,num_2)
            if num>0 and num<1750:
                list.append(num)
            else:
                print(num)
        list.append(1760-((1760-list[-1])/2))
        list.append(1760-((1760-list[-1])/2))
        print(list)
        for idx,i in enumerate(list):
            if idx ==0:
                time.sleep(1)
            currentMouseX,currentMouseY = pyautogui.position()
            if currentMouseY > 1060:
                keep = False
                pyautogui.press( 'esc' )
                break
            else:
                pyautogui.moveTo(i,1012)
            if not keep:
                break
            pyautogui.click(button = 'left')
            block = abs(random.gauss(mu=1, sigma=1))
            time.sleep(2+block)
            
        if not keep:
            break
        pyautogui.moveTo(160,1060)
        pyautogui.click(button = 'right')
```

```python
def on_press(key):
    # 按下按鍵時執行
    try:
        if key == keyboard.Key.space:
            pyautogui.moveTo(781,597)
            pyautogui.click(button = 'left' )
            print("開始執行")
            thread_tenderDeclaration_detail = threading.Thread(target=started)
            thread_tenderDeclaration_detail.start()
    except AttributeError:
        pass
    #通過屬性判斷按鍵類型。

def on_release(key):
    # 鬆開按鍵時執行
    try:
        print('{0} released'.format(key))
        if key == keyboard.Key.esc:
            global keep
            keep = False
            # Stop listener
            return False
    except AttributeError:
        pass
```

```python
for i in range(0,100):
    print("按下 space 開始執行")
    # Collect events until released
    with keyboard.Listener(
            on_press=on_press,
            on_release=on_release) as listener:
        listener.join()
```
